```python
import numpy as np
import matplotlib.pyplot as plt

R = 8.314  # J/(mol*K)

def calculate_Cp(T, a, b, c):
    return a + b * T + c / T**2

# Function to calculate H, S, and G
def calculate_properties(T, a, b, c, S298):
    Cp = calculate_Cp(T, a, b, c)
    
    # Enthalpy (H)
    H = S298 + np.cumsum(Cp)

    # Entropy (S)
    S = S298 + np.cumsum(Cp / T)

    # Gibbs free energy (G)
    G = H - T * S

    return H, S, G

# Constants for Argon
a_argon = 20.79  # J/(mol*K)
b_argon = 2.83e-7  # J/(mol*K)
c_argon = -3.66e-8  # J/(mol*K)
S298_argon = 154.8  # J/(mol*K)

# Constants for Titanium
a_Ti = 24.94  # J/(mol*K)
b_Ti = 6.57e3  # J/(mol*K)
c_Ti = -1.63e-5  # J/(mol*K)
S298_Ti = 30.75  # J/(mol*K)

# Constants for Titanium Dioxide
a_TiO2 = 73.35  # J/(mol*K)
b_TiO2 = 3.05e3  # J/(mol*K)
c_TiO2 = 17.03e-5  # J/(mol*K)
S298_TiO2 = 50.6  # J/(mol*K)

# Temperature range
temperature_range = np.linspace(298, 750, 250)  # K

H_argon, S_argon, G_argon = calculate_properties(temperature_range, a_argon, b_argon, c_argon, S298_argon)
H_Ti, S_Ti, G_Ti = calculate_properties(temperature_range, a_Ti, b_Ti, c_Ti, S298_Ti)
H_TiO2, S_TiO2, G_TiO2 = calculate_properties(temperature_range, a_TiO2, b_TiO2, c_TiO2, S298_TiO2)

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(temperature_range, H_argon - H_argon[0], label='Ar - (H)')
plt.plot(temperature_range, S_argon - S_argon[0], label='Ar - (S)')
plt.plot(temperature_range, G_argon - G_argon[0], label='Ar - (G)')

plt.plot(temperature_range, H_Ti - H_Ti[0], label='Ti -  (H)')
plt.plot(temperature_range, S_Ti - S_Ti[0], label='Ti -  (S)')
plt.plot(temperature_range, G_Ti - G_Ti[0], label='Ti -  (G)')

plt.plot(temperature_range, H_TiO2 - H_TiO2[0], label='TiO2 - (H)')
plt.plot(temperature_range, S_TiO2 - S_TiO2[0], label='TiO2 - (S)')
plt.plot(temperature_range, G_TiO2 - G_TiO2[0], label='TiO2 - (G)')

plt.xlabel('Temperature (K)')
plt.ylabel('Enthalpy (J/mol)')
plt.legend()
plt.title('Enthalpy, Entropy, and Gibbs Free Energy vs Temperature')
plt.grid(True)
plt.show()
```
